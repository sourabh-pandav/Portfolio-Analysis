```python


#Problem 1
import numpy as np
import pandas as pd

# Cash flows
cf = [0, 7, 7, 7, 7, 7, 7, 7, 7, 107]

# Rates
rates = [0.027, 0.029, 0.031, 0.035, 0.0375, 0.037, 0.0365, 0.037, 0.039, 0.043, 0.046]

# Step 1 - Discount factors
discount = np.cumprod([1] + [1/(1+r) for r in rates[:-1]])

# Step 2 - Zero rates
zero = [(1/d)**(1/i) - 1 for i,d in enumerate(discount, 1)]

# Step 3 - Par rates
par = [0] + [(1 - discount[i])/sum(discount[1:i+1]) for i in range(1,6)]
par.extend(rates[6:])

# Step 4 - Fill in curves
maturities = [i for i in range(len(rates))]
df_rates = pd.DataFrame({'Maturity':maturities, 'Input':rates,
                         'Discount':discount, 'Zero':zero, 'Par':par})

print("Discount, zero, and par rate curves:")
print(df_rates)

# Step 5 - PV of bond
pv = sum(cf[i]*discount[i] for i in range(len(cf)))
print(f"\nPV of bond: {pv:.2f}")

# Step 6 - Change forward curve and compute values
print("\nChanging each forward rate by +0.1%:")

dfs = []
for i in range(6, len(rates)):
  new_rates = rates.copy()
  new_rates[i] += 0.001

  new_discount = np.cumprod([1] + [1/(1+r) for r in new_rates[:-1]])
  new_pv = sum(cf[j]*new_discount[j] for j in range(len(cf)))

  temp_df = pd.DataFrame({'Changed Rate': [i], 'New PV': [new_pv]})
  dfs.append(temp_df)

shifted_df = pd.concat(dfs, ignore_index=True)
print(shifted_df)

# Step 7 - Compute duration
duration = sum(cf[i]*discount[i]*(i+1) for i in range(len(cf))) / pv
print(f"\nDuration: {duration:.2f}")

# Step 8 - Compute convexity
pv_down = sum(cf[i]*np.cumprod([1] + [1/(1+r-0.001) for r in rates[:-1]])[i] for i in range(len(cf)))
pv_up = sum(cf[i]*np.cumprod([1] + [1/(1+r+0.001) for r in rates[:-1]])[i] for i in range(len(cf)))
convexity = (pv_up + pv_down - 2*pv) / (pv * 0.001**2)
print(f"Convexity: {convexity:.2f}")

# Step 9 - Forward price
fwd_rate = zero[-1] # 30 month zero rate
fwd_price = pv * (1 + fwd_rate)**30 / discount[-1]
print(f"\nForward Price at 30 months: {fwd_price:.2f}")
```

```python
#Problem 2
import numpy as np
from scipy.stats import norm

# Inputs
S = 915 # Bond price
K = 900 # Strike price
vol = 0.05 # Volatility
r = 0.015 # Risk-free rate
T = 8.25 # Years to maturity
Tc = 0.25 # Years to coupon
C = 25 # Coupon amount

# Time to maturity and coupon
T = T - Tc

# Calculate d1 and d2
d1 = (np.log(S/K) + (r + 0.5*vol**2)*T) / (vol * np.sqrt(T))
d2 = d1 - vol*np.sqrt(T)

# N(d1) and N(d2)
N_d1 = norm.cdf(d1)
N_d2 = norm.cdf(d2)

# Call price
C_price = S*N_d1 - np.exp(-r*Tc)*K*N_d2

# PV of coupon
C_pv = C * np.exp(-r*Tc)

# Quoted call price
C_quoted = S*N_d1 - np.exp(-r*Tc)*K*N_d2 - C_pv

# Put-call parity for put price
P_price = C_price + K*np.exp(-r*T) - S

print("Call Price:", C_price)
print("Quoted Call Price:", C_quoted)
print("Put Price:", P_price)

# Compare strike to cash and quoted prices
print("\nStrike vs. Cash Price:", K < S)
print("Strike vs. Quoted Price:", K < (S - C_pv))
```

```python
#Problem 3
import numpy as np
import matplotlib.pyplot as plt

# Yields
maturities = [0.25, 0.5, 1, 2, 3, 5, 7, 10, 20]
rates = [0.012, 0.0123, 0.013, 0.0155, 0.0187, 0.024, 0.0314, 0.0345, 0.038]

# Cubic spline fit
A = np.zeros((len(maturities), len(maturities)))
A[0,0] = 1
A[-1,-1] = 1
for i in range(1, len(maturities)-1):
  A[i,i-1] = maturities[i] - maturities[i-1]
  A[i,i] = 2 * (maturities[i+1] - maturities[i-1])
  A[i,i+1] = maturities[i+1] - maturities[i]

b = np.zeros(len(maturities))
for i in range(len(maturities)-1):
  b[i] = (rates[i+1] - rates[i]) / (maturities[i+1] - maturities[i])

x = np.linalg.solve(A, b)

# Print coefficients
print("Cubic Spline Coefficients:")
for i in range(len(x)):
  print(f"d{i+1}: {x[i]:.6f}")

# Evaluate spline
years = np.arange(0, 20.1, 0.1)
spline_rates = np.zeros_like(years)
for i in range(len(maturities)-1):
  spline_rates += x[i] * (years - maturities[i])**3

# Print rates
print("\nYears\tSpline Rates")
for i in range(len(years)):
  print(f"{years[i]:.1f}\t{spline_rates[i]:.6f}")

# Plot
plt.plot(years, spline_rates)
plt.plot(maturities, rates, 'o')
plt.title("Yield Curve - Cubic Spline Fit")
plt.xlabel("Years")
plt.ylabel("Yield")
plt.show()

# Rate at 4 years
print(f"\nRate at 4 years: {spline_rates[np.where(years==4)[0][0]]:.6f}")
```

```python
#problem 4
import numpy as np
from scipy.optimize import minimize

# Treasury bond data
times = [0.4356, 0.2644, 0.2658, 0.4342, 0.0192, 0.4753, 0.3534, 0.1, 0.2685, 0.4342,
         0.2274, 0.1027, 0.2712, 0.437, 0.4822, 0.226, 0.4822, 0.226, 0.2301, 0.4808,
         0.4932, 0.4959, 0.2397, 0.4959, 0.2397, 0.4959, 0.2397, 0.2438, 0.4945]

maturities = [0.4356, 0.7644, 1.2658, 1.9342, 2.0192, 2.9753, 3.3534, 3.6, 4.2685, 4.9342,
              5.2274, 5.6027, 6.2712, 6.937, 7.4822, 7.726, 8.4822, 8.726, 9.2301, 9.9808,
             25.4932, 26.4959, 26.7397, 27.4959, 27.7397, 28.4959, 28.7397, 29.2438, 29.9945]

coupon_rates = [0.0078, 0.0078, 0.0065, 0.0053, 0.0028, 0.0065, 0.014, 0.0165, 0.0203,
                0.0165, 0.044, 0.0228, 0.0265, 0.0228, 0.034, 0.0378, 0.0265, 0.0303,
                0.0328, 0.0253, 0.044, 0.0465, 0.049, 0.0428, 0.044, 0.034, 0.0415,
                0.0438, 0.0378]

prices = [100.3, 100.48, 100.5, 100.31, 99.78, 100.16, 102.34, 103.08, 104.19, 102.06,
          115.91, 104.36, 105.86, 102.97, 110.53, 113.09, 103.98, 106.5, 108, 101.19,
          117.58, 122.28, 126.97, 115.19, 117.47, 98.98, 112.44, 114.67, 105.75]

T = np.array(maturities)

# Initial guesses
b0 = 0.02
b1 = 0.01
b2 = 0.01
lamb = 0.1
params = np.array([b0, b1, b2, lamb])

# Nelson-Siegel model
def model(params, T):
  b0, b1, b2, lamb = params

  rates = b0 + b1*(1 - np.exp(-lamb*T))/T + b2*((1 - np.exp(-lamb*T))/T - np.exp(-lamb*T))
  return rates

# Objective function
def objective(params):
  pred = model(params, T)
  return np.sum((pred - prices)**2)

# Optimization
opt = minimize(objective, params, method='Nelder-Mead')

# Calculate yields
b0, b1, b2, lamb = opt.x
y = model(opt.x, T)

# Print results
print("Beta_0:", b0)
print("Beta_1:", b1)
print("Beta_2:", b2)
print("Lambda:", lamb)
print("Sum:", objective(opt.x))

# Yields and weights
# Table output
print("{:<10} {:<10} {:<10} {:<10} {:<10} {:<10} {:<10} {:<10} {:<10} {:<10}".format(
      'Time', 'Freq', 'Maturity', 'Coupon', 'Price', 'Interest', 'Dirty Price', 'Yield', 'B_hat', 'Weight'))

for i in range(len(T)):
  ti = T[i]
  Bi = prices[i]
  ci = coupon_rates[i]
  yi = y[i]

  interest = ci*Bi*ti
  dirty_price = Bi + interest
  Bi_hat = Bi / (1 + yi*ti)
  wi = (Bi - Bi_hat)**2

  print("{:<10.4f} {:<10} {:<10.4f} {:<10.4f} {:<10.2f} {:<10.6f} {:<10.6f} {:<10.7f} {:<10.6f} {:<10.6f}".format(
        ti, 2, T[i], ci, Bi, interest, dirty_price, yi, Bi_hat, wi))

# Graph
import matplotlib.pyplot as plt

plt.plot(T, y)

# Add data points
plt.scatter(T, y, c='r', marker='o')

plt.title("Yield Curve")
plt.xlabel("Time to Maturity")
plt.ylabel("Yield")
plt.show()
```
